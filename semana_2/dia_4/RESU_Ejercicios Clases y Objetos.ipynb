## Ejercicio 1
Somos una empresa que tiene varias tiendas de electrodomesticos, con ventas en todo el territorio nacional. Necesitamos un programa para manejar las tiendas, ver las ventas que han tenido, cuántos empleados hay, etc... Para ello vamos a modelizar la tienda en una clase, que tendrá las siguientes características:
* Nombre clase: "Tienda"
* Atributos comunes:
    * Negocio: "Electrodomésticos"
    * Abierta: True
* Atributos propios de cada tienda:
    * Nombre - String
    * Dirección - String
    * Número de empleados - int
    * Ventas ultimos 3 meses - Lista de 3 numeros
* Método para calcular las ventas de todos los meses, que devuelva un numero con todas las ventas.
* Método que calcula la media de ventas de los ultimos meses, por empleado
* Método que devuelve en un string el nombre de la tienda, junto con su dirección.
* Método que saca las ventas del último mes.
* Método para dar una proyección de las ventas en caso de haber invertido X dinero en marketing en esa tienda. Siendo X el parámetro de entrada. Si X es menor de 1000, las ventas de los ultimos 3 meses hubiesen sido de (1.2 x ventas), si es mayor o igual de 1000, las ventas hubiesen sido de (1.5 x venta). El método reescribe el atributo de ventas en los últimos 3 meses, y además devuelve las ventas con el aumento.

Se pide:
1. Implementar la clase "Tienda"
2. Crear tres tiendas con datos inventados
3. Comprueba en al menos una de ellas, todo lo implementado en la clase tienda (sus atributos, media de ventas, ventas/empleado...)
3. Calcular las ventas del último mes de todas las tiendas. Para ello usa el bucle `for`
4. Imprime por pantalla los nombres de las tiendas cuya dirección lleve el string "Avenida"
5. Seguro que ya has hecho este apartado... Documenta la clase :)

```python
# Apartado 1
class Tienda:
    '''
    Clase con los componentes de una tienda de Electrodomesticos S.A.
    atributos:
        negocio = Electrodomestico -> String
        abierta = True -> Bool
    '''
    negocio = "Electrodomésticos"
    abierta = True
    
    def __init__(self, nombre, direccion, num_empleados, ventas_3_meses):
        '''
        Elementos particulares de cada tienda
        Input:
            nombre: string
            direccion: string
            num_emp: int
            ventas_3_meses: list
        '''
        self.nombre = nombre
        self.direccion = direccion
        self.num_empleados = num_empleados
        self.ventas_3_meses = ventas_3_meses
        
    def tot_ventas(self):
        '''
        Metodo que suma todas las ventas de la tienda
        Output:
            total de ventas -> float/int
        '''
        return sum(self.ventas_3_meses)
    
    def vent_empl(self):
        '''
        Calcula la media de de las ventas por empleado
        output:
            float con la media
        '''
        
        # Opcion 1
        #return sum(self.ventas_3_meses)/self.num_empleados
        
        # Opcion 2
        #vent_tot = sum(self.ventas_3_meses)
        #ratio_empl = vent_tot/self.num_empleados
        #return ratio_empl
        
        # Opcion 3
        return self.tot_ventas()/self.num_empleados
    
    def nombre_direc(self):
        '''
        Metodo que devuelve un string con el nombre y la direccion
        Output:
            string
        '''
        return "Nombre tienda: " + self.nombre + "\nDireccion: " + self.direccion
    
    def ventas_ult_mes(self):
        '''
        Ventas del utlimo mes
        Output:
            float/int
        '''
        v_ult_mes = self.ventas_3_meses[-1]
        return v_ult_mes
    
    def proyec_marketing(self, marketing):
        '''
        Método para dar una proyección de las ventas en caso de haber invertido X dinero 
        en marketing en esa tienda. Siendo X el parámetro de entrada. Si X es menor de 1000, 
        las ventas de los ultimos 3 meses hubiesen sido de (1.2 x ventas), si es mayor o igual de 1000,
        las ventas hubiesen sido de (1.5 x venta). El método reescribe el atributo de ventas en los 
        últimos 3 meses, y además devuelve las ventas con el aumento
        Input:
            marketing: cantidad invertida en marketing -> float/int
        Output:
            lista de floats con el resultado de aplicar el algoritmo marketing
        '''
        output = []
        
        for i in self.ventas_3_meses:
            if marketing < 1000:
                output.append(i*1.2)
                
            else:
                output.append(i*1.5)
                
        self.ventas_3_meses = output
        return output
```

## Ejercicio 2
En este ejercicio vamos a implementar una clase *Perro* en Python. La clase tiene las siguientes características:
* Cosas que sabemos seguro que tiene un perro
    * Tiene 4 patas
    * 2 orejas
    * 2 ojos
    * Una velocidad de 0. Por defecto, el perro está parado
* Cuando se inicialice:
    * El perro será de una determinada raza
    * Por defecto tendrá pelo "Marrón", a no ser que se diga lo contrario.
    * Por defecto no tendrá dueño, a no ser que se diga lo contrario.
    
* Dispondrá también de un método llamado andar, que tiene un argumento de entrada (aumento_velocidad). Este valor se le sumará a la velocidad que ya llevaba el perro.
* Necesita otro método (parar), donde pondremos la velocidad a 0.
* Otro método llamado "ladrar", que tendrá un argumento de entrada, y la salida será el siguiente string: "GUAU!" + el argumento de entrada.


Se pide:
* Implementa la clase *Perro*
* Crea un objeto de tipo *Perro*, sin dueño
* Comprueba que están bien todos sus atributos
* Prueba que ande, y comprueba su velocidad
* Páralo
* Documenta la clase *Perro*

```python
class Perro:
    '''
    Clase perro de ejemplo
    Un perro tiene estos atributos:
        num_patas = 4 -> int
        num_orejas = 2 -> int
        num_ojos = 2 -> int
        velocidad = 0 -> int
    '''
    num_patas = 4
    num_orejas = 2
    num_ojos = 2
    velocidad = 0
    
    def __init__(self, raza, pelo = "Marrón", duenio = None):
        '''
        Los perros se diferencian unos de otros por los siguientes atributos
        inputs constructor:
            raza -> string
            pelo -> string ("Marrón" por defecto)
            duenio -> string (None por defecto)
        '''
        self.raza = raza
        self.pelo = pelo
        self.duenio = duenio
        
    def andar(self, aumento_velocidad):
        '''
        Aumenta el atributo velocidad
        Input:
            aumento_velocidad -> float
        Output: None
        '''
        #self.velocidad = self.velocidad + aumento_velocidad
        self.velocidad += aumento_velocidad
        
    def parar(self):
        '''
        Ponemos el atributo velocidad a 0
        Output: None
        '''
        self.velocidad = 0
        
    #def set_velocidad(self, velocidad):
     #   self.velocidad = velocidad
        
    def ladrar(self, mensaje):
        '''
        Método para mandar un mensaje
        Input: mensaje -> string
        Output: string
        '''
        return "GUAU!, " + mensaje
```

```python
san_ber = Perro("San Bernardo")
print(san_ber.raza)
print(san_ber.pelo)
print(san_ber.duenio)

print(san_ber.velocidad) # velocidad = 0

san_ber.andar(2)
print(san_ber.velocidad) # velociada = 2

san_ber.andar(2)
print(san_ber.velocidad) # velocidad = 4

san_ber.parar()
print(san_ber.velocidad) # velocidad = 0

san_ber.ladrar("quiero comer")
```
